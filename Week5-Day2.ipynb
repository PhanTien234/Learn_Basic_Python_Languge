```python
from random import randint

height = [randint(58,78) for x in range(20)]
weight = [randint(90,250) for x in range(20)]

plt.scatter(weight,height)
plt.title("Height-Weight Distribution")
plt.xlabel("Weight(lbs)")
plt.ylabel("Height(inches)")
plt.show()
```

```python
import numpy as np

np.random.seed(5)
ages=[np.random.normal(loc=40,scale=10) for x in range(1000)]
plt.hist(ages,bins=45)
plt.title("Ages per Population")
plt.xlabel("Age")
plt.ylabel("# of People")
plt.show()
```

```python
florida = [np.random.normal(loc=60,scale=15) for x in range(1000)]
california = [np.random.normal(loc=35,scale=5) for x in range(1000)]

#red chart
plt.hist(florida,bins=45,color="r",alpha=0.4)
#blue chart
plt.hist(california,bins=45,color="b",alpha=0.4)

plt.show()
plt.hist(california,bins=45,color="g",alpha=1)
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd

movieData = pd.read_csv('Moviedata.csv')
movieData.columns=['Film','Genre','CriticRatings','AudienceRatings','Budget','Year'] #Change Column Name
df = movieData.groupby("Year", as_index=False).sum().head()

x=df.Year
y=df.Budget

plt.plot(x,y)
plt.title("Budget Usage for Movies")
plt.xlabel("Year")
plt.ylabel("Total Budget(Millions)")
plt.show()
```

```python
df =movieData.groupby("Genre",as_index=False).sum().head(100)

num_people = df.Budget
categories = df.Genre

plt.bar(categories,num_people)
plt.title("Movie Category")
plt.xlabel("Category")
plt.ylabel("Budget Usage(Millions)")
plt.show()
```

```python
df =movieData.groupby("Genre",as_index=False).sum().head(100)
budget = df.Budget
category = df.Genre
plt.scatter(category,budget)
plt.title("Budget-Category Distribution")
plt.xlabel("Category")
plt.ylabel("Total Budget (millions)")
plt.show()
```

```python
df = movieData.groupby("Year",as_index=False).mean().head(100)

rating=df.AudienceRatings
year=df.Year
budget=df.Budget

plt.scatter(year,rating, c=budget)
plt.title("Movie Rating and Budget Over Time")
plt.xlabel("Year")
plt.ylabel("Audience Ratings")
plt.colorbar(label="Budget(Million)")
plt.grid(True)
plt.show()
```
